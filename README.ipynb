## 一，课题背景

近年来，人们对通过外场操纵粒子扩散的实验研究越来越感兴趣。 通过改变场参数可以有效地控制扩散过程，通过改变外部电磁场来增加离子在膜通道中的扩散率成为可能，可以通过改变不同方向的扩散系数有效地分离不同性质的微米和纳米粒子[3]尽管取得了这些显著成就，但迄今为止，对这种在外力影响下增强的扩散的定量描述仍然不完整，而且对于活性布朗粒子的研究更是少只有少。 本研究有助于这一主题，重点是含时周期势场下的过阻尼活性粒子的扩散与整流。

关于棘轮效应

棘轮效应[5]是指在受驱动的非对称周期势系统中发生定向粒子输运的现象，在噪声或波动的辅助下，没有任何明显的外部偏压。 棘轮效应在解释生物和物理世界的不同过程中的成功应用导致了理论和实验研究的激增[8]、[9]、[10]、[11]、[12]、[13]。 虽然对棘轮的主要研究旨在解释生物或分子马达的工作原理[14]，但目前它已扩展到其他各个领域

**棘轮效应**，在各个学科中都有出现，我这里讲得是，在不对称势下的定向移动，

## 模型介绍

我们考虑了过阻尼非相互作用活性布朗粒子在二维管道中的扩散，自驱动速度为$v_0$，含时周期势场是$V(x)=-A\cos(\omega t+k x)$,A是周期势场中的振幅，$\omega,k$是其频率和波数，粒子所处于二维无限长管道中，

这是过阻尼郎之万方程：，在x方向上受到，自驱动力，和行波作用力，和热噪音，管道如图所视，y方向长度为a，x方向无限长。我分别弹性和粘性边界条件为都试了下，后面的结果一样，关于扩散系数用的是这个公式，初始状态是沿着x方向向右运动

## 结果讨论

结果非常有意思，图一作为对比我方在这里了，纵坐标是$log(r^2)$，这是理论求出来的不加外场的活性粒子的扩散，短时间是次扩散，长时间也是正常扩散，中间有一段是超扩散。图二是加上外场后，明显开看出了，短时间是次扩散，长时间是指数为2的超扩散，图三是扩散中心沿着行波的方向匀速前进，不过为为这里主要研究的是短时间（5秒）内的次扩散。

图一的横坐标是不同的$\omega$,纵坐标是模拟出的扩散系数，不同颜色代表了不同的自驱动速度也就是活性，我们可以看出来，在一定范围内，活性高的粒子，扩散系数反而降低，图二是补充了更多的粒子，不是很明显，我们分别减去他们的理论值，就是图三，我们可以；明显看出来，在低频情况下活性反而会抑制粒子的扩散，而高频情况下，曲线逼近0，不影响扩散。，图四，是每一个自驱动速度的最佳扩散系数对应的频率随着自驱动速度的增加呈指数增加

另外我们还可以发现，粒子的运动会在行波的波谷振荡，形成明暗相见的pattern，我先补充一个机器学习的算法，基于密度进行聚类的无监督算法，DBSCAN，
 这个算法核心思想就是划圆圈；随即取一个点，以epsilon为半径画圆圈，如果圈内的点数大于设置的MinPts，这些点将会被视为一个簇，否则，视为噪音点，然后遍历这个圈内的点继续画圆，扩充这个簇

我用DBSCAN算法对x坐标进行动态聚类，追踪，pattern数，，但是这个结果长时间模拟是不正确的，因为需要由于扩散的进行，epsilon需要动态改变，但是为之前写的程序里epsilon的值是个定值，所以长时间模拟的结果还需要调整以下epsilon
